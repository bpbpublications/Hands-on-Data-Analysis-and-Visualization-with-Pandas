# Introduction to Matplotlib 

## Artistic Layer

```python
import pandas as pd
import numpy as np
from matplotlib.backends.backend_agg import FigureCanvas
from matplotlib.figure import Figure

fig = Figure()
canvas = FigureCanvas(fig)
ax = fig.add_subplot(111)
# Generate 10000 random variables
a = np.random.randn(10000)
# Assign the random variables to a series object x
x = pd.Series(a)

# Plot the variables using histogram method with bin size=60
ax.hist(x, bins=60)
# set Label for X - axis
ax.set_xlabel("x-Axis")
# set Label for Y - axis
ax.set_ylabel("y-Axis")
# set Label for Title
ax.set_title("Normal distribution graph")
# creates an image file normaldistribution.png
canvas.print_png("normaldistribution.png")
```

## Figure and Axes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
x = np.array([2, 4, 6, 8])
fig = plt.figure(figsize=(4, 4), facecolor="grey")
ax = fig.add_axes([0, 0, 0.6, 0.4])
ax.plot(x, x ** 3)
ax.set_title("Cubes")
ax1 = fig.add_axes([0.8, 0, 0.8, 0.6])
ax1.plot(x, x ** 2)
ax1.set_title("Squares")
```

## Formating Line type and color

```python
import matplotlib.pyplot as plt

x = np.array([1, 2, 3])
y = np.array([2, 4, 6])
plt.plot(x, y, "--", c="b", label="Dash Line")
plt.plot(x, x, "-", c="g", label="Solid Line")
plt.plot(x, x * x, "-.", c="r", label="Dash-dotted")
plt.plot(x, x / 2, ":", c="m", label="Dotted")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Matplotlib - Formatting")
plt.legend()
```

## Adding Annotations

```python
import matplotlib.pyplot as plt

year = [2015, 2016, 2017, 2018, 2019, 2020]
pizza_prizes = [300, 420, 520, 700, 864, 1064]
plt.plot(year, pizza_prizes, "o-")
plt.xlabel("Year")
plt.ylabel("Pizza prizes")
plt.title("Pizza prizes india 2015-2020")
# Annotation
plt.annotate(xy=[2015, 300], s="Price Low")
plt.annotate(xy=[2019, 864], s="Price High")
```

## Legends

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y1, "-b", label="sine")
plt.plot(x, y2, "--r", label="cosine")
plt.legend(loc=10)
```

## Line plots

```python
import matplotlib.pyplot as plt

Year = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010]
india_Unemployment_Rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3]
brazil_Unemployment_Rate = [8.6, 10, 9, 11, 14, 13, 14, 12.2, 13.6, 15.0]

plt.plot(Year, india_Unemployment_Rate, Label="India", color="r", marker="o")
plt.plot(Year, brazil_Unemployment_Rate, Label="Brazil", color="b", marker="o")

plt.title(" India Vs Brazil Unemployment 2001-2010", fontsize=14)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Unemployment Rate", fontsize=14)
plt.grid(True)
plt.legend()
```

## Pie Charts

```python
import matplotlib.pyplot as plt

slices = [7, 2, 6, 13]
activities = ["playing", "eating", "working", "sleeping"]
cols = ["g", "y", "r", "c"]
plt.pie(
    slices,
    labels=activities,
    colors=cols,
    startangle=90,
    shadow=True,
    explode=(0.2, 0.0, 0.0, 0.1),
    autopct="%1.1f%%",
)
plt.title("What i do on sundays?")
plt.show()
```

## Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

fig, a = plt.subplots(2, 2)
x = np.arange(1, 10)
a[0][0].plot(x, x * x)
a[0][0].set_title("square")
a[0][1].plot(x, x * x * x, c="m")
a[0][1].set_title("cube")
a[1][0].plot(x, x * x * x * x, c="y")
a[1][0].set_title("quad")
a[1][1].plot(x, np.exp(x), c="r")
a[1][1].set_title("exponential")
plt.subplots_adjust(left=0, bottom=0, right=1.0, top=1.0, hspace=0.27)
```
