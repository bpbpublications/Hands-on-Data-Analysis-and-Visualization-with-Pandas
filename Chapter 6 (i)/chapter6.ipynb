## Handling Large Datasets

```python
import pandas as pd
import numpy as np
```

```python
%%time
df = pd.read_csv('datasets/sales_data.csv', parse_dates=['Ship Date', 'Order Date'])
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.info(memory_usage="deep")
```

```python
for dtype in ["float", "int", "object", "datetime"]:
    df_dtype = df.select_dtypes(dtype)
    mean_usage_in_bytes = df_dtype.memory_usage(deep=True).mean()
    mean_usage_in_megabytes = mean_usage_in_bytes / 1024 ** 2
    print(
        "Average memory usage for {} columns: {:03.2f} MB".format(
            dtype, mean_usage_in_megabytes
        )
    )
```

## Datatype subtypes

```python
int_sub_types = ["uint8", "int8", "int16", "int64"]
for subtype in int_sub_types:
    print(np.iinfo(subtype))
```

### Reduce the Integer types

```python
def dtype_mem_usage(obj):
    if isinstance(obj, pd.DataFrame):
        result = obj.memory_usage(deep=True).sum()
    else:
        result = obj.memory_usage(deep=True)
    result = result / 1024 ** 2
    return "{:2.2f} MB".format(result)
```

```python
df_int = df.select_dtypes(include=["int"])
to_uint = df_int.apply(pd.to_numeric, downcast="unsigned")
```

```python
print(dtype_mem_usage(df_int))
print(dtype_mem_usage(to_uint))
```

### The Int64 reduced by 62 percent using unassigned dtype

### Reduce the Float data types

```python
df_float = df.select_dtypes("float")
to_float = df_float.apply(pd.to_numeric, downcast="float")
```

```python
print(dtype_mem_usage(df_float))
print(dtype_mem_usage(to_float))
```

### The Float64 reduced by 50 percent using float32

```python
optimised_df = df.copy()
```

```python
optimised_df[to_uint.columns] = to_uint
optimised_df[to_float.columns] = to_float
```

```python
optimised_df.info()
```

```python
print(dtype_mem_usage(df))
print(dtype_mem_usage(optimised_df))
```

### Optimising the int64 and float64 values reduced by 7 percent 

### Working  with categorical variables

```python
df_object = df.select_dtypes(include=["object"]).copy()
```

```python
df_object.Region.head()
```

```python
df_region = df_object.Region
df_region_cat = df_region.astype("category")
```

```python
df_region_cat.head(5).cat.codes
```

```python
df_region_cat.head(5)
```

```python
to_object = pd.DataFrame()
for x in df_object.columns:
    unique_values = len(df_object[x].unique())
    total = len(df_object[x])
    if unique_values / total < 0.5:
        to_object.loc[:, x] = df_object[x].astype("category")
    else:
        to_object.loc[:, x] = df_object[x]
```

```python
print(dtype_mem_usage(df_object))
print(dtype_mem_usage(to_object))
```

### Converting object to category have reduced from object to category by 98 percent which is greatetst achievement

```python
optimised_df[to_object.columns] = to_object
dtype_mem_usage(optimised_df)
```

```python
optimised_df.info(memory_usage="deep")
```

```python
print("Memory usage of the main Dataframe  : {} ".format(dtype_mem_usage(df)))
print(
    "Memory usage of the Optimised Dataframe  : {} ".format(
        dtype_mem_usage(optimised_df)
    )
)
```
